### Nominatim API query

#### Geopy alternative

```python
con = mysql.connector.connect(host="localhost", user="root", password="", database="dataengineer",)
cursor = con.cursor()

sql_query = "SELECT * FROM address"
test = pd.read_sql(sql_query, con)

test['location'] = test['address'].apply(lambda x: ' '.join([x.split(' ')[0], x.split(' ')[-1]]))
test['location'] = test['location'] + ', ' + test.city + ', ' + test.postal_code + ', FRANCE'
```

```python
geolocator = Nominatim(user_agent="my_request") # Create instance of Nominatim class
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1) # Wrapped instance with query rate limitation

try:
    test['geocode'] = test['location'].apply(geocode) # Applying geocode method to get location into DataFrame

except Exception as inst:
    print(inst)
```

```python
test.geocode.isna().sum(), test.shape
```

```python
test['Lat'] = test['geocode'].apply(lambda x: x.latitude if x else '')
test['Lon'] = test['geocode'].apply(lambda x: x.longitude if x else '')

test.drop(['location', 'geocode'], axis=1, inplace=True)
#test = test.fillna("")
```

```python
test.to_csv('address_geocode.csv', index=False, sep=';', encoding='utf-8')  
```

#### URl alternative

```python
essai = test.copy()
postal_code = essai["postal_code"] + " " + essai["city"] + " France"

lat = []
lon = []

for p in postal_code:
        url = "https://nominatim.openstreetmap.org/search?q=" +p+"&limit=1&format=json"
        responses = requests.get(url) 
        responsejs = responses.json()
        
        if len(responsejs) != 0:
            lat.append(responsejs[0]['lat'])
            lon.append(responsejs[0]['lon'])
        else:
            lat.append("")
            lon.append("")

            
essai['Lat']= lat
essai['Lon'] = lon

essai["Lat"] = pd.to_numeric(essai["Lat"])
essai["Lon"] = pd.to_numeric(essai["Lon"])
```

```python
essai.isna().sum()
```

### SQL update output

```python
essai = essai.fillna("")
val_update = essai[['address_id', 'Lat', 'Lon']].values.tolist()

try:
    con = mysql.connector.connect(host="localhost", user="root", password="", database="dataengineer")
    cursor = con.cursor()
    
    query = 'UPDATE address SET latitude=%s, longitude=%s WHERE address_id=%s;'
    cursor.executemany(query, [(x[1], x[2], x[0]) for x in val_update])
    con.commit()
    print("Rows updated = ", cursor.rowcount)
    
finally:
    cursor.close()
    con.close()
```

### SQL backup dump

```python
HOST = 'localhost' 
USER = 'root'
PWD = ''
DB_NAME = 'dataengineer'
DB_backup = 'dataengineer_updated'
PATH = os.getcwd() + '/backup/'

DATE = time.strftime('%Y%m%d-%H%M%S')
backupPPATH = PATH + DATE

try:
    os.stat(backupPPATH)
    
except:
    os.mkdir(backupPPATH)


dumpcmd = "mysqldump -h " + HOST + " -u " + USER + " -p" + PWD + " " + DB_NAME + " > " + pipes.quote(backupPPATH) + "/" + DB_backup + ".sql"
os.popen(dumpcmd)

os.system(dumpcmd)
gzipcmd = "gzip " + pipes.quote(backupPPATH) + "/" + DB_backup + ".sql"
os.system(gzipcmd)

print ("Backup created in '" + backupPPATH + "' directory")
```
